```
def ordenar_dos_tipus(llista):
    j = length = len(llista) - 1
    i = 0
    while i < j:
        while i < length and not(llista[i].isalpha()): 
            i += 1
        while j > 0 and llista[j].isalpha(): 
            j -= 1
        if i < j:
            llista[i], llista[j] = llista[j], llista[i]

llista = ["456","789","abc","123","tzu","870","zzz"]
ordenar_dos_tipus(llista)
llista
```
