# 5.1 Suma d'una llista

**Digues quin és el cas base de l'algorisme i calcula'n la seva complexitat.**

    El cas base és quan la llista té només un o cap element. La complexitat de l'algorisme és O(n):
    Tenim un problema de mida n, que reformulem mitjançant la solució de 2 problemes de mida n/2 i llavors combinem
    les respostes en un temps O(n^0).
    Per tant estem al cas 3 del Teorema Master: O(n^log_b a) = O(n) en aquest cas.

```
def suma_llista(llista):
    #Casos base
    if len(llista) == 1:
        return llista[0]
    
    if len(llista) == 0:
        return 0
    
    else:
        mig = len(llista) // 2
        right = llista[:mig]
        left = llista[mig:]
        return suma_llista(right) + suma_llista(left)
```

```
assert suma_llista([1,2,3,4]) == 10
assert suma_llista([-1,-2,0,1,2]) == 0
assert suma_llista([4,2,-1,-3]) == 2
assert suma_llista([-1,2,-1,1,-2]) == -1
assert suma_llista([1,-2,3,4]) == 6
assert suma_llista([0]) == 0
assert suma_llista([]) == 0
```

# 5.10 Comptador inversions en una llista

```
def comptador_inversions(llista):
    numInversions = 0
    for i in range(len(llista)):
      for j in range(i+1, len(llista)):
        if (llista[i] > llista[j]):
            numInversions += 1
            
    return numInversions
```

# 5.22 Ordenar llista aparellada

**Calcula la complexitat d’aquests dos algoritmes i mesura el temps d’execució del teu programa per cadascun.**

    -->Complexitat amb mergesort:
    Quan dividim creem 2 problemes, de mida n/2. I per combinar les solucions, tenim d = 1.
    Per tant, com que a=2 i b=2, log_b a = 1, que és igual a d, podem aplicar el cas 2 del teorema Master:
    mergesort té una complexitat O(n log n).

    -->Complexitat amb quicksort:
    Si totes les particions passen al mig del vector tenim el millor cas, i el nombre de comparacions serà:
    T(n) = 2 T(n/2) + n, que segons el teorema Master és O(n log n).

    En el pitjor cas (p.e. [4,3,2,1,0]), totes les particions són als extrems (alguna de les subllistes estarà buida), 
    llavors el nombre de comparacions serà T(n) = O(n^2).

    Observació: Sempre podem desordenar la llista al principi i aquest cas no existirà mai.

    En el cas promig el nombre de comparacions serà O(1,38 n log_2 n), és a dir, un 38% més de comparacions que en el
    millor cas.

    La complexitat se seguiria expressant com a O(n log n).

```
#Mergesort
def ordenar_parelles(llista):
    if len(llista) > 1:
        mig = len(llista) // 2
        left = llista[:mig]
        right = llista[mig:]

        ordenar_parelles(left)
        ordenar_parelles(right)

        i = 0
        j = 0
        k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                llista[k] = left[i]
                i += 1
            else:
                llista[k] = right[j]
                j += 1
            k += 1
        
        while i < len(left):
            llista[k] = left[i]
            i += 1
            k += 1
        
        while j < len(right):
            llista[k] = right[j]
            j += 1
            k += 1
```

```
llista = [3,3,1,1,5,5,0,0]
ordenar_parelles(llista)
assert llista == [0,0,1,1,3,3,5,5]

llista = [1,1,4,4,3,3,5,5]
ordenar_parelles(llista)
assert llista == [1, 1, 3, 3, 4, 4, 5, 5]

llista = [2,2,1,1,6,6,8,8,9,9,0,0]
ordenar_parelles(llista)
assert llista == [0, 0, 1, 1, 2, 2, 6, 6, 8, 8, 9, 9]

llista = [-1,-1,-3,-3,0,0,1,1,3,3,4,4]
ordenar_parelles(llista)
assert llista == [-3, -3, -1, -1, 0, 0, 1, 1, 3, 3, 4, 4]

import time
%timeit(ordenar_parelles([-1,-1,-3,-3,0,0,1,1,3,3,4,4]))
```

```
#Quicksort
def ordenar_parelles2(llista):
    quick_sort(llista, 0 , len(llista) - 1)

def quick_sort(llista, first, last):
    if last > first:
        pivot = partition(llista, first, last)
        quick_sort(llista, first, pivot - 1)
        quick_sort(llista, pivot + 1, last)

def partition(llista, first, last):
    mid = (first + last) // 2
    if llista[first] > llista[mid]:
        llista[first], llista[mid] = llista[mid], llista[first]
    if llista[first] > llista[last]:
        llista[first], llista[last] = llista[last], llista[first]
    if llista[mid] > llista[last]:
        llista[mid], llista[last] = llista[last], llista[mid]
    llista[mid], llista[first] = llista[first], llista[mid]

    pivot = first
    i = first + 1
    j = last 
    
    while True:
        while i <= last and llista[i] <= llista[pivot]:
            i += 1
        while j >= first and llista[j] > llista[pivot]:
            j -= 1
        if i >= j:
            break
        else:
            llista[i], llista[j] = llista[j], llista[i]
    llista[j], llista[pivot] = llista[pivot], llista[j]
    return j
```

```
llista = [3,3,1,1,5,5,0,0]
ordenar_parelles2(llista)
assert llista == [0,0,1,1,3,3,5,5]

llista = [1,1,4,4,3,3,5,5]
ordenar_parelles2(llista)
assert llista == [1, 1, 3, 3, 4, 4, 5, 5]

llista = [2,2,1,1,6,6,8,8,9,9,0,0]
ordenar_parelles2(llista)
assert llista == [0, 0, 1, 1, 2, 2, 6, 6, 8, 8, 9, 9]

llista = [-1,-1,-3,-3,0,0,1,1,3,3,4,4]
ordenar_parelles2(llista)
assert llista == [-3, -3, -1, -1, 0, 0, 1, 1, 3, 3, 4, 4]

import time
%timeit(ordenar_parelles2([-1,-1,-3,-3,0,0,1,1,3,3,4,4]))
```
