```
def preguntar_nom ():
  nom = input ("Com et dius?")
  edat = eval(input("Quina edat tens?"))
  print ("Hola", nom, "tens", edat, "anys")

preguntar_nom()
```

```
def preguntar_pais ():
  pais = input ("De quin país ets?")
  ciutat = input ("De quina ciutat ets?")
  edat = eval(input("Quina edat tens?"))

  print ("Ets de", pais)
  print ("I de la ciutat", ciutat)
  print ("Ets de", pais + ", i de la ciutat", ciutat, "i tens", edat, "anys.")

preguntar_pais ()
```
