# Exercici 5.6 Negatius al davant

```
def negatius(llista):
    for i in range(len(llista)):
        index  = i
        for j in range(i+1, len(llista)):
            if llista[index] > llista[j]:
                index = j
        llista[i], llista[index] = llista[index], llista[i]
```

```
llista = [1, -2, 3, -4, -3, 5, 6]
negatius(llista)
assert llista == [-4, -3, -2, 1, 3, 5, 6]
llista = [-2, 0, 1, 14, -3, -1, -2]
negatius(llista)
assert llista == [-3, -2, -2, -1, 0, 1, 14]
llista = [0, 1, -2, -1, -1, -1, 0]
negatius(llista)
assert llista == [-2, -1, -1, -1, 0, 0, 1]
llista = [10, 1, -11, -2, 4, 1, -0]
negatius(llista)
assert llista == [-11, -2, 0, 1, 1, 4, 10]

# Aquest resultat és un d'entre molts possibles, # la única condició és que els negatius
# estiguin al principi de la llista.
```

# Exercici 5.19 Karatsuba

**Complexitat operacions implicades:**

    Per multiplicar dos nombres de n bits (x i y), l'algorisme de Karatsuba fa tres multiplicacions i algunes adicions,
    i tracta nombres que son aproximadament la meitat del tamany original de x i y.

    Utilitzant el Teorema Master tenim un problema de mida n, que reformulem mitjançant la solució de a (a=3)
    problemes de mida n/b (b=2), i llavors combinem les respostes (les adicions que fa) en un temps O (n^d) (d=1).

    Llavors la complexitat és de O(n^log_2(3)).
    Perquè d < log_b (a).



**Quantes vegades es crida a l'executar karatsuba(1000,1050).**
    
    El primer cop es crida amb (0, 50) en r1: tenim com a màxim 4, valor mig 2, a = 10, b = 0, c = 10, d = 50. I retorna 0 (cas base i r1)
    El segon cop es crida amb (10, 60) en r2, d'aquí es crida un tercer cop amb (0,0) en r1. I retorna 0 (cas base).
    D'aquí també es crida un quart cop amb (1,6) en r2, i retorna 6 (cas base).
    Es torna a crida un cinqué cop amb (1, 6) per r3, i retorna 6 (cas base). 
    D'aquest apartat tenim r1 = 0,  r2 = 6 i r3 = 6 i llavors prod = 600. (r2 de la primera crida).
    Es crida un sisè cop amb (10,10) per r3 de la primera crida:
    d'aquí es crida un setè cop amb (0,0) per r1 i retorna 0 (cas base).
    Es torna a cridar (vuitè cop) amb (1, 1) per r2, i retorna 1 (cas base).
    Fa la novena crida també amb (1,1) per r3, i retorna 1 (cas base).
    D'aquí ens queda r1 = 0, r2 = 1, r3 = 1 i llavors prod = 100 (r3 de la primera crida).

    Finalment, el return prod, fruit de la primera crida ens retorna el resultat de 1000 * 1050 = 1050000.



```
def karatsuba(x,y):
    if x < 10 or y < 10:
        return x*y  #Cas base   #O(1)

    maxim = max(len(str(x)), len(str(y)))
    mig = maxim // 2

    a = x // 10**(mig)
    b = x % 10**(mig)
    c = y // 10**(mig)
    d = y % 10**(mig)

    r1 = karatsuba(b, d)
    r2 = karatsuba((a+b), (c+d))
    r3 = karatsuba(a, c)

    prod = (r3 * 10**(2*mig)) + ((r2 - r3 - r1) * 10**(mig)) + (r1)

    return prod
```

```
assert karatsuba(1100,4050) == 4455000
assert karatsuba(12312,12) == 147744
assert karatsuba(9232,2112) == 19497984
assert karatsuba(2,12) == 24
```

# Exercici 5.23 Sumatori parcial màxim (dividir i vèncer)

**Demostra que la complexitat d’aquest algorisme és O(n log(n)).**

    En el cas base, n = 1 és la llargada de la llista, i això vol dir que podem executar l'algorisme amb O(1).

    Per buscar el sumatori parcial màxim necessitem O(n). Hem de cridar recursivament amb la part esquerra i dreta
    de la llista, i per buscar el sumatori parcial de cadascuna de les llistes necessitem O(n/2) perquè les hem
    dividit per la meitat. Com que tenim 2 subllistes, la complexitat total (temps) d'aquestes subllistes serà 2T(n/2).

    Utilitzant el Teorema Master tenim un problema de mida n, que reformulem mitjançant la solució de a (a=2) problemes de mida n/b (b=2),
    i llavors combinem les respostes en un temps O (n^d) (d=1).

    La complexitat llavors és O(n log(n)) ja que d = log_b a:

    d = 1
    b = 2
    a = 2
    log_2 (2) = 1 = d

```
def sumatori_parcial_divider_vencer(llista):
    if len(llista) == 1:  #Cas base
        return llista[0]
    
    #Dividim la llista per la meitat.
    mig = len(llista) // 2

    max_left = float('-inf')
    acum_left = 0

    #Sumatori màxim de la primera meitat de la llista (des de la meitat fins l'inici)
    for i in range(mig-1, -1, -1):    
        acum_left += llista[i]
        if max_left < acum_left:
            max_left = acum_left

    max_right = float('-inf')
    acum_right = 0

    #Sumatori màxim de la segona meitat de la llista (des de la meitat fins al final)
    for j in range(mig, len(llista)):
        acum_right += llista[j]
        if max_right < acum_right:
            max_right = acum_right
    
    #Valor del sumatori parcial màxim de la primera meitat (crida recursiva)
    rec_esquerra = sumatori_parcial_divider_vencer(llista[:mig])  
    
    #Valor del sumatori parcial màxim de la segona meitat (crida recursiva)
    rec_dreta = sumatori_parcial_divider_vencer(llista[mig:])
    max_sum = max(max_right + max_left, rec_esquerra, rec_dreta)

    return max_sum
```

```
assert sumatori_parcial_divider_vencer([-3, 1, -5, 2, 7, 8]) == 17
assert sumatori_parcial_divider_vencer([1, 2, -5, 3, 6, -2, 4]) == 11
assert sumatori_parcial_divider_vencer([0, -1, -5, -1, 6, 1, 9]) == 16
assert sumatori_parcial_divider_vencer([0, 0, -5, -1, 0, 0, 0]) == 0
assert sumatori_parcial_divider_vencer([-4, -3, -5, -1, -2, -1, -1]) == -1
assert sumatori_parcial_divider_vencer([1, 2, -6, 4, -1, 2, 1, -5]) == 6
assert sumatori_parcial_divider_vencer([-3, 1, -5, 2, 7, 8]) == 17
```

# Exercici 5.23 Sumatori parcial màxim (Kadane)

```
def sumatori_parcial_kadane(llista):
    max_sum = float('-inf')   #suma màxima que hem trobat fins al moment
    acum = 0    #suma parcial de tots els elements anteriors

    for i in range(len(llista)):    #recorrem la llista una única vegada.
        if acum < 0:    #la suma parcial esdevéé negativa, descartem la suma parcial i reiniciem el valor de la suma parcial.
            acum = 0
        acum += llista[i]
        if (max_sum < acum):
            max_sum = acum
    
    if max_sum < 0:   #Cas en que tota la llista sigui de nombres negatius.
        max_sum = abs(max_sum)
    return max_sum
```

```
assert sumatori_parcial_kadane([1, 2, -5, 3, 6, -2, 4]) == 11
assert sumatori_parcial_kadane([0, -1, -5, -1, 6, 1, 9]) == 16
assert sumatori_parcial_kadane([0, 0, -5, -1, 0, 0, 0]) == 0
assert sumatori_parcial_kadane([-4, -3, -5, -1, -2, -1, -1]) == 1
assert sumatori_parcial_kadane([1, 2, -6, 4, -1, 2, 1, -5]) == 6
assert sumatori_parcial_kadane([-3, 1, -5, 2, 7, 8]) == 17
```
