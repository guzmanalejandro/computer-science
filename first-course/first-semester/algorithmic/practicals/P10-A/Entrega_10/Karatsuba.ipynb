```
def zeroPad(numberString, zeros, left = True):
    """Retorna un string amb zeros afegits a la dreta o esquerra."""
    if left:
        numberString = '0'*zeros + numberString
    else:
        numberString = numberString + '0'*zeros
    return numberString

def karatsuba(x,y):
    x = str(x)
    y = str(y)
    if len(x) == 1 and len(y) == 1:
        return int(x) * int(y)
    if len(x) < len(y):
        x = zeroPad(x, len(y) - len(x))
    elif len(y) < len(x):
        y = zeroPad(y, len(x) - len(y))
    n = len(x)
    m = n // 2
    # per enters senars
    if (n % 2) != 0:
        m += 1  
    BZeroPadding = n - m
    AZeroPadding = BZeroPadding * 2
    a = int(x[:m])
    b = int(x[m:])
    c = int(y[:m])
    d = int(y[m:])
    ac = karatsuba(a, c)
    bd = karatsuba(b, d)
    k = karatsuba(a + b, c + d)
    A = int(zeroPad(str(ac), AZeroPadding, False))
    B = int(zeroPad(str(k - ac - bd), BZeroPadding, False))
    return A + B + bd

karatsuba("234","5")
```
