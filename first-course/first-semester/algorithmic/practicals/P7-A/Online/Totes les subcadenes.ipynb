1. Escriu una funció que retorni una llista amb totes les subcadenes d’una cadena donada. Fes servir list comprehensions.
Per exemple: totes_subcadenes("abcd") hauria de retornar la llista ['a', 'ab', 'abc', 'abcd', 'b', 'bc', 'bcd', 'c', 'cd', 'd']

```
def totes_subcadenes(cadena):
    subcadenes = [cadena[init : end+1] for init in range(len(cadena))
    for end in range(init, len(cadena))]
    return subcadenes
```

```
totes_subcadenes('noah')
```

```
assert len(totes_subcadenes('abcd')) == int((len('abcd') * (len('abcd') + 1)/2))
assert totes_subcadenes('abcd') == ['a', 'ab', 'abc', 'abcd', 'b', 'bc', 'bcd', 'c', 'cd', 'd']
```
