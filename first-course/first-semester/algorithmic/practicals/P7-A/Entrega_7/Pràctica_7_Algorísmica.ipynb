# Exercici 4.1: Acrònims

1. Escriu una funció, **acro**, que a partir d’una frase sense punts, comes, accents, número ni cap signe de puntuació, imprimeixi l’acrònim corresponent.




```
def acronim(frase):     #O(n^2)
    separada = frase.split()    #O(n)
    separada = [separada.capitalize() for separada in separada]   #O(n^2)
    acro = [separada[0] for separada in separada] #O(n^2)
    return "".join(acro)    #O(1)
```

```
assert acronim("Hola que tal") == 'HQT'
assert acronim("Hola em dic Marina") == 'HEDM'
assert acronim("Tic divuit anys") == 'TDA'
assert acronim("Vaig de vacances") == 'VDV'
assert acronim("Administrador de Infraestructuras Ferroviarias") == 'ADIF'
```

# Exercici 4.6: Run Length Encoding

1. Escriu una funció **rle** que donat un text de lletres ASCII (A-Z), el codifiqui usant l’algorisme RLE.


```
def rle(text):      #O(n*m) --> O(n) de mitja, O(n*m) el pitjor dels cassos, amb N com la llargada de la cadena més llarga i M de la més curta q busquem.
    text_codificat = ''   #O(1)
    comptador = 1         #O(1)
    anterior_caracter = ''    #O(1)
    
    for caract in text:   #O(n*m)
      if caract != anterior_caracter:   #O(1)
        if anterior_caracter:           #O(1)
          text_codificat += anterior_caracter + str(comptador)   #O(1)
        comptador = 1     #O(1)
        anterior_caracter = caract    #O(1)
      else:
        comptador += 1        #O(1)
    
    text_codificat += anterior_caracter + str(comptador)  #O(1)  #Perquè retorni amb l'última lletra comptada també.
    return text_codificat     #O(1)
```

```
assert rle('ABBBBNNNEEEDDDZZAAAAA') == 'A1B4N3E3D3Z2A5'
assert rle('BBBBBBBBBBBBBBBBBWWWWWWZAAA') == 'B17W6Z1A3'
assert rle('AAAAABBBBBBCCCCCDDDDDEEEEEEAAAAAFFFFFF') == 'A5B6C5D5E6A5F6'
assert rle('PPPPPOOOOOOOOQQQQQQQQQQNNNNNNNRE') == 'P5O8Q10N7R1E1'
assert rle('WWWWWWWAACCDDEEFFGG') == 'W7A2C2D2E2F2G2'
assert rle('SSSDDDFFFGGGAAABBBBNNNNMMMMLLLL') == 'S3D3F3G3A3B4N4M4L4'
```

# Exercici 4.7: Subcadena més llarga sense cap caràcter repetit

Escriu una funció anomenada **subcadena_mes_llarga** que donada una cadena de caràcters, identifiqui la subcadena més llarga que no contingui cap caràcter repetit.

Per exemple, per a la cadena 'lacadenamesllarga', la subcadena més llarga sense cap caràcter repetit és 'namesl'.


```
def subcadena_mes_llarga(cadena):   #O(n)
    ult_aparicio = {}     #O(1)
    max_len = 0           #O(1)
    max_pos = 0           #O(1)
    actual_pos = 0        #O(1)
    actual_len = 0        #O(1)

    for comptador, valor in enumerate(cadena): #O(n) per la complexitat de la funció enumerate.
      l = ult_aparicio.get(valor, -1)   #O(n) pitjor dels casos, O(1) normalment.
      if l < actual_pos:     #O(1)
        actual_len +=1       #O(1) 
      else:
        if actual_len > max_len:    #O(1)
          max_pos = actual_pos      #O(1)
          max_len = actual_len      #O(1)        #Totes operacions simples.
        actual_len -= l - actual_pos    #O(1)
        actual_pos = l + 1              #O(1)
      ult_aparicio[valor] = comptador   #O(n) en el pitjor dels casos
    
    if actual_len > max_len:    #O(1)
      max_pos = actual_pos      #O(1)
      max_len = actual_len      #O(1)
    
    return cadena[max_pos: max_pos + max_len] #O(n) en el pitjor dels casos
```

```
assert subcadena_mes_llarga('lacadenamesllarga') == 'namesl'
assert subcadena_mes_llarga('mesllarga') == 'mesl'
assert subcadena_mes_llarga('aaa') ==  'a'
assert subcadena_mes_llarga('holacomestas') == 'lacomest'
assert subcadena_mes_llarga('pizzauvapomapera') == 'uvapom'
assert subcadena_mes_llarga('dillunsdimartsdimecresdijous') == 'lunsdimart'
```
