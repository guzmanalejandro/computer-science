Calcula el terme 10 i el terme 100:

Què ha passat quan has intentat calcular el terme 100? Per què?

Triga molt a donar la resposta degut a que al ser un algorisme recursiu, es va cridant a ell mateix, i en el cas del nombre 100, ha de fer moltes crides fins a donar una resposta.

```
def fib1(n):
    if n==0:
        return n
    if n==1:
        return n
    else:
        return fib1(n-1) + fib1(n-2)

fib1(10)
```

**Algorisme de Fibonacci amb una versió basada en llistes:**

Només executa (n-1) vegades la iteració. Direm que fib2(n) és **lineal** (o polinòmic) respecte de n. El seu únic defecte és que crea una llista que, per termes molt grans, serà molt gran.

```
def fib2(n):
  if n == 0:
    return 0
  ls = [0 , 1]
  for i in range(2, n+1):
    ls.append(ls[i-1] + ls[i-2])
  return ls[n]

fib2(1000)
```

**Un algorisme eficient per calcular els termes de la sèrie de Fibonacci**

Només executa (n-1) vegades la iteració. Direm que fib2(n) és **lineal** (o polinòmic) respecte de n. Però ara té un valor afegit: la quantitat de memòria que fa servir és mínima.

```
def fib3(n):
    a,b = 0,1
    for i in range(1,n+1):
        a,b = b, a+b
    return a

fib3(10000)
```
