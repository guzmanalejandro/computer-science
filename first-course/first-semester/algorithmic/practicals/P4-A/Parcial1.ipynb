### Pregunta 1
Explica de forma raonada perquè els algorismes no depenen del llenguatge en que estan escrits, o dir d'una altra manera, perquè el mateix algorisme escrit en dos llenguatges pot tenir un aspecte molt diferent.

**Resposta**: Els algorismes són independents tant del llenguatge de programació en que s'expressen com de l'ordinador que els executa. A cada problema que es presenti, l'algoritme es pot expressar en un llenguatge diferent de programació i executar-se en un ordinador diferent; però l'algorisme serà sempre el mateix. Això és perquè un algorisme són les idees/estrategies que hi ha darrera dels programes per tal de solucionar un problema.

### Pregunta 2
Explica de forma clara què vol dir que un algorisme és **correcte** i **eficient**.

**Resposta**: Un algorisme correcte és aquell que aporta una solució correcta al nostre problema, en altres paraules, un algorisme capaç de solucionar el nostre problema. Que un algorisme sigui correcte (que solucioni el nostre problema) no implica que sigui eficient. Entenem per eficiencia la quantitat de recursos utilitzats per un algorisme, i sempre tractarem de buscar una eficiencia màxima, per tal de minimitzar l'ús dels recursos.

### Pregunta 5

Siguin $m$ i $n$ dos nombres enters. Es diu que $m$ divideix $n$ si i només si existeix un valor $k$ tal que $n = k \times m$.

Completa tots aquells llocs d'aquest programa que tinguin `...` de manera que calculi els divisors d'un nombre `num`:

```python
def divisors(num):
    divisors= []
    for divisor in ...:
        if num%divisor == 0:
            divisors.append(divisor)

```

````python
def divisors(num):
    divisors= []
    for divisor in range(1, num + 1):
        if num%divisor == 0:
            divisors.append(divisor)
    return divisors

divisors(15)
````
