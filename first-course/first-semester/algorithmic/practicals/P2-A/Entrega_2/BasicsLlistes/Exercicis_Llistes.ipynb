Exercici 2: Llista selectiva

Escriu una funci√≥ que donat un nombre  ùëõ  i una llista, retorni una segona llista amb els elements de la llista inicials que s√≥n estrictament menors al nombre  ùëõ  donat.

Exemple:

llistaSelectiva([1, 15, 5, 3, 2, 12, 99, 8, 6], 10)

[1, 5, 3, 2, 8, 6]

```
def llistaSelectiva(list, n):
  newList = []
  for j in list:
    if j < n:
      newList.append(j)
  return newList

llistaSelectiva([1, 15, 5, 3, 2, 12, 99, 8, 6], 10)
```

Introdueix un n√∫mero: -99

S'han introduit 3 nombres, amb una suma total de 12, i una mitjana aritm√®tica de 4

```
def introduirNombres():
    llista = []
    x = int(input("Introdueix un n√∫mero (-99 per acabar): "))
    while x != -99:
        llista.append(x)
        x = int(input("Introdueix un n√∫mero (-99 per acabar): "))
    nombres = len(llista)
    suma = sum(llista)
    mitja = int(suma / nombres)
    print("S'han introduit", nombres, "nombres, amb una suma total de"
    ,suma, ", i una mitjana aritm√®tica de",mitja)

introduirNombres()
```

Consulta la documentaci√≥ de Python per si hi ha funcions est√†ndards que et puguin ajudar.

Exemple:

Introdueix un n√∫mero: 7

Introdueix un n√∫mero: 6

Introdueix un n√∫mero: -1

Introdueix un n√∫mero: -99

El n√∫mero m√†xim introduit √©s el 7 i el m√≠nim el -1

```
def maxmin():
    llista = []
    xStr = eval(input("Introdueix un n√∫mero (-99 per acabar): "))
    while xStr != -99:
        llista.append(xStr)
        xStr = eval(input("Introdueix un n√∫mero (-99 per acabar): "))
    maxim = max(llista)
    minim = min(llista)
    print("El n√∫mero m√†xim introduit √©s el", maxim, "i el m√≠nim el", minim)

maxmin()
```

Exercici 8: Inici i final d'una llista

Escriu un programa que donada una llista en retorni una de nova amb nom√©s el primer i el darrer element de la llista.

Exemple:

iniciFinal([3, 6, 1, 9, 10])

[3, 10]

```
def iniciFinal(llista):
  return [llista[0], llista[-1]]

iniciFinal([3, 6, 1, 9, 10])
```

Exercici 9: Canviar les paraules d'ordre

Escriu una funci√≥ que donada una frase la retorni a l'usuari amb les paraules en ordre invers.

Exemple:

Em dic Andreu

Andreu dic Em

```
def flippedWords(frase):
  return " ".join(frase.split()[::-1])

flippedWords('Em dic Andreu')
```
