# 3.1. Divisó entera

1. Escriu una funció que faci la divisió entera entre dos nombres enters, positius i més grans que 1. No utilitzis els operadors *, /, % per a resoldre el problema.

**Reflexions prèvies**: Quantes operacions fa l’algorisme quan calculem 10/2?

    op1: 10 - 2 // quocient(0) + 1
    op2: 8 - 2 // quocient (1) + 1
    op3: 6 - 2 // quocient (2) + 1
    op4: 4 - 2 // quocient (3) + 1
    op5: 2-2 // quocient (4) + 1
    RESULTAT: 5 (10 OPERACIONS)

```
def divisio_entera(dividend, divisor):
    
    if dividend == 0:    #Casos extrems en que el dividend o el divisor sigui 0.
      return 0
    if divisor == 0:
      raise ValueError("No es pot dividir per 0.") 

    if dividend < 0:    #Tractant els números negatius.
      dividend = - dividend
      
      if divisor < 0:
        divisor = - divisor
        resultatNegatiu = False
      else:
        resultatNegatiu = True
    
    elif (divisor < 0):
      divisor = - divisor
      resultatNegatiu = True
    
    else:
      resultatNegatiu = False

    quocient = 0    #Fem ús d'un comptador 'quocient' que va incrementant el seu
    while (dividend >= divisor):                #valor fins que dividend < divisor.
      dividend = dividend - divisor
      quocient += 1
    
    if (resultatNegatiu):
      quocient = - quocient
    return quocient
```

```
assert divisio_entera(10,2) == 5
assert divisio_entera(9,2) == 4
assert divisio_entera(-10,-2) == 5
assert divisio_entera(25,-5) == -5
assert divisio_entera(0,2) == 0
```

# 3.2.El Màxim Comú Divisor i les seves aplicacions

Fent servir l’algorisme anterior, escriu una funció que simplifiqui una fracció a la seva expressió irreductible. Recorda que una fracció irreductible és una fracció en la qual el numerador i el denominador no tenen cap divisor comú.

**Reflexions prèvies**: Fes aquestes reflexions abans de començar a programar:


*   Explica els passos que fa l’algorisme en el cas reduir_fraccio(12, 8).

    Per reduir la fracció amb numerador 12 i denominador 8, el primer que fem és cridar a la funció amb el codi de Euclides per a calcular el màxim comú divisor de dos nombres, en aquest cas de 12 i de 8, això ens dóna com a resultat: 4.

    Com que una fracció irreductible és la que el numerador i el denominador no tenen cap divisor en comú, dividim aquest 4 primer al numerador, i desprès al denominador, obtenint així la fracció irreductible: 3/2.

*   Calcula el cost de l’algorisme per reduir_fraccio(127,63).

    En dues iteracions successives els dos arguments decreixen almenys a la meitat, és a dir, perden un bit en la seva representació. Si inicialment eren enters de n bits, en 2n iteracions arribarem al final de l'algorisme. Com que cada iteració implica una divisió (b mod a) d'ordre quadràtic O(n^2), el temps total serà O(n^3).


```
#Codi Euclides, per trobar el màxim comú divisor de dos nombres a i b:
def gcd(a,b):
        while a:
            a,b = b%a, a
        return b

def reduir_fraccio(numerador, denominador):
    mcd = gcd(numerador, denominador)
    numReduit = numerador // mcd
    denReduit = denominador // mcd
    
    return numReduit, denReduit
```

```
assert reduir_fraccio(12, 8) == (3,2)
assert reduir_fraccio(6, 8) == (3,4)
assert reduir_fraccio(4, 6) == (2,3)
assert reduir_fraccio(26, 8) == (13,4)
```

# 3.5. Operacions amb nombres binaris

1. Determinar si un nombre és potència de 2 fent ús de les operacions binàries tot i seguint les següents indicacions:


*   Si ens fixem en la representació binària d’un nombre és fàcil veure que les potències de 2 només tenen un bit a 1.
*   A nivell de bits, donada una potència de 2 (p.e. 010000), el nombre anterior té una forma complementaria en els bits menys significatius (001111).

**Reflecions prèvies**: Fes aquestes reflexions abans de començar a programar:


*   Revisa les operacions amb booleans. Quin algorisme aplicarem?

    El primer terme comprova si és una potència de 2 assegurant-se que tots els bits desprès d'aquesta operació 'bitwise &' són 0. L'operació de 'bitwise' està dissenyada per ser només True per a potències de 2.


*   Explica els passos que fa potencia2(8)

    Per a potencia2(8) amb l'operador bitwise AND comprovem que 8(num) & 7(num-1) == 0, i això és cert perquè per bitwise &(dissenyada per ser només True per a potències de 2), per a cada bit en el qual les dues entrades són 1, la sortida és 1. En canvi, si un o els dos bits són 0, la. sortida és 0 per al bit. El número 8 en binari és 0b1000 i el número 7 és 0b111, per això al fer l'operació de bitwise & es compleix que sigui == 0. Comprovem també que el num introduït és diferent de 0, i llavors ens retorna True.


*   Quins casos especials hem de tenir en compte?

    Python té nombres enters amb una precissió arbitrària, llavors això funciona per a qualsevol enter n sempre que 'capiga' a la memòria. L'operació de 'bitwise' està dissenyada per ser només True per a potències de 2, amb una excepció: si num(i, per tant, tots els seus bits) fossin 0 al començar.

```
def potencia2(num):
    es_potencia = (num & (num-1) == 0) and num !=0
    return es_potencia
```

```
assert potencia2(1024) == True
assert potencia2(2**2345) == True
assert potencia2(2**2345-1) == False
assert potencia2(0) == False
assert potencia2(2**50) == True
assert potencia2(512) == True
```
