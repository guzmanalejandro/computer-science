# Exercici 5.8 Rotacions

```
def rotacions(llista):
    num_rotacions = 0
    llargada = len(llista)
    for i in range(llargada):
        if llista[llargada-1] < llista[i]:
            llista[i] = llista[llargada-1]
            num_rotacions += 1 
    return num_rotacions
```

```
assert rotacions([9, 10, 2, 5, 6, 8]) == 2
assert rotacions([3, 5, 10, 12, 1, 2]) == 4
assert rotacions([20, 2, 3, 7, 15, 18]) == 1
assert rotacions([1, 2, 3, 4, 6, 8]) == 0
assert rotacions([4, 5, 0, 1, 2, 3]) == 2
assert rotacions([19, 1, 2, 3, 15, 16]) == 1
```

# Exercici 5.23 Sumatori parcial màxim

```
def sumatori_parcial_maxim_forca_bruta(llista):
    llista = [llista[i:j+1] for i in range(len(llista)) for j in range(i, len(llista))]
    max_sum = 0

    for i in llista:
        suma = sum(i)
        if suma > max_sum:
            max_sum = suma

    return max_sum
```

```
assert sumatori_parcial_maxim_forca_bruta([1, 2, -6, 4, -1, 2, 1, -5]) == 6
assert sumatori_parcial_maxim_forca_bruta([3, 1, -8, -4, 0, 9, 1, 5]) == 15
assert sumatori_parcial_maxim_forca_bruta([1, 2, 4]) == 7
assert sumatori_parcial_maxim_forca_bruta([3, 2, -1, 2, 6, 1, 1, 1]) == 15
assert sumatori_parcial_maxim_forca_bruta([3, 0]) == 3
```

# Exercici 5.24 Xifres i lletres

**Demostra que la complexitat de l’algorisme que has escrit és O(n2).**

El funcionament de l'algorisme es basa en una seqüència d'operacions que repeteix fins a acabar d'ordenar una llista:

-->Primer buscar el minim element de la llista.

-->Després el canvia amb el primer.

-->Busca el següent mínim en la resta de la llista.

-->L'intercanvia amb el segon.

En resum, l'algorisme busca el mínim element entre una posició 'i' i el final de la llista i intercanvia el mínim amb l'element de la posició 'i'.

L’operació més important és una comparació:

if llista[ind_minim] > llista[j]: ind_minim = j

I el nombre de vegades que s’executa és:

C(n) = n + (n-1) + (n-2) + ... + 2 que acaba sent O(n^2).

L'algorisme és quadràtic, O(n^2), tot i que només fem O(n) intercanvis a la llista. 

```
def xifres_lletres_ord_sel(llista):
    for i in range(len(llista)):
        ind_minim = i
        for j in range(i+1, len(llista)):
            if llista[ind_minim] > llista[j]:
                ind_minim = j
        llista[i], llista[ind_minim] = llista[ind_minim], llista[i]
```
