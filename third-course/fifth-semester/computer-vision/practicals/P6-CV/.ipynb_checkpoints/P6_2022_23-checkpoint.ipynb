**1.1** Build a function `to_integral_image` that computes the integral image of an input (2D) array. The integral image must have an additional row and column full of zeros (first row and first column).
Make sure that the values of the integral image are correct.

Try your function using a `5x5` random grayscale image. Visualize both the original random image and the integral one.

```python
#your solution
def to_integral_image(image):
    if(len(image.shape) >= 3):
        image = rgb2gray(image)
    image = image.astype(np.int64)
    x, y = image.shape
    tmp = np.arange(x * y).reshape((x, y))

    def integral_sum(j):
        i = j//x
        j = j%x
        if(i!=0)&(j!=0):
            image[i][j] += image[i][j-1] + image[i-1][j] - image[i-1][j-1]
        elif(j!=0):
            image[i][j] += image[i][j-1]
        elif(i!=0):
            image[i][j] += image[i-1][j]

    np.vectorize(integral_sum)(tmp)
    
    new_image = np.zeros((x+1, y))
    new_image[1:,:]=image
    return new_image
```

```python
grayscale_img = np.random.randint(0, 255, (5, 5))
ii_grayscale_img = to_integral_image(grayscale_img)

plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1)
plt.imshow(grayscale_img, cmap='gray')
plt.title('random grayScale')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(ii_grayscale_img, cmap='gray')
plt.title('integral image')
plt.axis('off')
```

## Principal Component Analysis (PCA) applied for face recognition
==============================================================================================

### Dimensionality and redundancy

Imagine we have a dataset with 100x100 pixel images, it means we have 10000 dimensions. We want to construct a low-dimensional linear subspace that best explains the variation in the set of face images (**Eigenfaces space**)

<img src="notebook_images/subspace.PNG" width=200, height=200>

Each image has m rows and n columns and defines a vector of (mxn) elements. We need to choose the most valuable pixels in order to avoid compute all dimensions. 

<img src="notebook_images/feature_vector.PNG" width=800, height=400>

We look for a transformation of the original space to a smaller (M << (mxn)) where faces are represented with their coordinates in this new space R.

To reduce the dimensionality retaining the information necessary to classify and recognize, we are going to use the **Eigenfaces method** 

### How to build a reduced space?

To build this new space, we are going to use the **Principal Component Analysis**. Given a large space, the PCA looks for the minimum number of axes that best represents the variation of the data.

<img src="notebook_images/pca.PNG" width=400, height=400>

The eigenvectors of the covariance matrix define the axis of maximum variance and the eigenvalues give a measure of the variance of the data. 

1. Construct the vector in the (m x n)-dimensional space R given M images of size (m x n).

2. Compute the mean image 

<center>
$\overline{X}=\frac{1}{M}\sum_{i=1}^{M} X_i$
</center>

3. Construct the covariance matrix. Due to $A \times A^T$ is too large, instead of using $A \times A^T$ to compute its eigenvectors, we are going to compute the eigenvectors of $A^T \times A$.

<img src="notebook_images/covariance_image.PNG" width=1500, height=300>

4. Extract the eigenvectors (the base of the new space) and their eigenvalues and project faces in the new space to apply the classifier (knn)

## 5. Compute PCA

[*class sklearn.decomposition.PCA(n_components=None, copy=True, whiten=False, svd_solver='auto', tol=0.0, iterated_power='auto', random_state=None)*](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html?highlight=pca#sklearn.decomposition.PCA)

The principal components measure deviations about this mean along orthogonal axes.

**5.1** Create a PCA object, using the training set and a 150 components

```python
pca = PCA(n_components=150)
```

```python
#your solution
n_components = 150

print(
    "Extracting the top %d eigenfaces from %d faces" % (n_components, X_train.shape[0])
)
t0 = time()
pca = PCA(n_components=n_components, svd_solver="randomized", whiten=True).fit(X_train)
print("done in %0.3fs" % (time() - t0))

eigenfaces = pca.components_.reshape((n_components, h, w))

print("Projecting the input data on the eigenfaces orthonormal basis")
t0 = time()
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
print("done in %0.3fs" % (time() - t0))
```

Plot the accumated variance of the components. 

**Hint:** Use the returned `explained_variance_ratio_`.

```python
explained_variance_ratio = pca.explained_variance_ratio_

cumulative_sum = np.cumsum(explained_variance_ratio)

plt.plot(cumulative_sum)
plt.xlabel('Number of components')
plt.ylabel('Cumulative sum of explained variance')
plt.show()
```

Please, explaing what is:
- accuracy (score)
- precision
- recall
- f1-score
- support
- macro avg
- weighted avg?


accuracy (score): the fraction of samples that were correctly classified. This metric is calculated as the number of correct predictions divided by the total number of samples.

precision: the fraction of predicted positive samples that are actually positive. This metric is calculated as the number of true positive predictions divided by the total number of positive predictions.

recall: the fraction of positive samples that were correctly predicted. This metric is calculated as the number of true positive predictions divided by the total number of positive samples.

f1-score: the harmonic mean of precision and recall. This metric is calculated as the product of precision and recall divided by their sum.

support: the number of samples for each class.

macro avg: the average of the evaluation metrics across all classes. This metric is calculated by taking the average of the metric values for each class, without considering the relative frequencies of the classes.

weighted avg: the weighted average of the evaluation metrics across all classes. This metric is calculated by taking the average of the metric values for each class, weighting each value by the relative frequency of the class.

* Computing time can change when using PCA, as it involves performing singular value decomposition (SVD) on the training set, which can be computationally expensive. However, the computing time will depend on the size of the training set and the number of components specified in the PCA. In general, using fewer components will reduce the computing time, as it reduces the dimensionality of the data.

* Whether using PCA or not will give better results will depend on the specific dataset and the performance of the classifier without PCA. In some cases, using PCA can improve the performance of the classifier, as it reduces the dimensionality of the data and removes noise and redundant features. In other cases, using PCA may not have a significant effect on the performance of the classifier.

* The result of using PCA may change if the number of components is changed. Using a larger number of components may capture more information from the data, potentially improving the performance of the classifier. However, using too many components may also result in overfitting, where the classifier performs well on the training set but poorly on the test set.

* The result of using Adaboost may change if the number of estimators is changed. Using a larger number of estimators may improve the performance of the classifier, as it allows the model to fit the data more accurately. However, using too many estimators may also result in overfitting, where the classifier performs well on the training set but poorly on the test set. It is important to tune the hyperparameters of the classifier, such as the number of estimators, to find the optimal configuration that gives the best performance on the test set.

```python
n_components_list = [10, 25, 50, 200, 500]

for n_components in n_components_list:
    pca = PCA(n_components=n_components)

    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    adaboost = AdaBoostClassifier()
    adaboost.fit(X_train_pca, y_train)

    accuracy = adaboost.score(X_test_pca, y_test)
    print(f'Accuracy with {n_components} components: {accuracy:.4f}')

    y_pred = adaboost.predict(X_test_pca)
    print(classification_report(y_test, y_pred))
```

This attribute returns the percentage of variance explained by each component, and the cumulative sum of these percentages will give you the total amount of variance explained by a certain number of components.

* The number of components that gives the best result will depend on the specific dataset and the performance of the classifier. You can compare the evaluation metrics, such as accuracy, precision, and recall, to see which number of components gives the best performance.

* Yes, the computing time may change when using a different number of components in PCA. Using a larger number of components will typically increase the computing time, as it requires performing singular value decomposition (SVD) on a larger matrix.
